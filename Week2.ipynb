## Membuat bentuk Geometri 2D dari OpenCV

````python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
````

## Menambahkan Bentuk ke dalam Gambar Sesuai Koordinat

````python
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
````

````python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
````

````python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
````

````python
cv.circle(black_img,center=(400,400),radius=15,color=(255,0,0),thickness=75)
plt.imshow(black_img)
````

## Menambahan Garis

````python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
````

## Menambahkan Text

````python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
````

## Pembuatan Black Image Kembali

````python
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
````

## Inisialisasi NumPy Array

````python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
````

## Melakukan reshape pada Array

````python
pts = vertices.reshape((-1,1,2))
pts
````

## Penambahan Polyline

````python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
````

### 5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

````python
import random

img_boxes = np.zeros((512, 512, 3), dtype=np.uint8)

for _ in range(5):
    pt1 = (random.randint(0, 400), random.randint(0, 400))
    pt2 = (pt1[0] + random.randint(50, 100), pt1[1] + random.randint(50, 100))
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    cv.rectangle(img_boxes, pt1=pt1, pt2=pt2, color=color, thickness=-1)

img_random_boxes = cv.cvtColor(img_boxes, cv.COLOR_BGR2RGB)

plt.imshow(img_random_boxes)
plt.show()
````

### 6. Tampilkan image dengan posisi terbalik!

````python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(img_random_boxes)
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(cv.flip(img_random_boxes, 0))
plt.title('Vertical')

plt.subplot(1, 3, 3)
plt.imshow(cv.flip(img_random_boxes, 1))
plt.title('Horizontal')


plt.show()
````

### 8. Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!  

````python
img3 = cv.imread('/content/drive/MyDrive/PCVK/Modul 2/Images/lena.png')
img3 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)

R, G, B = cv.split(img3)

y, w, h = B.shape[0] - 60, 50, 50

cv.rectangle(B, (10, y), (10 + w, (y+h)), (255), -1)
img3 = cv.merge([R, G, B])
plt.axis('off')
plt.imshow(img3)
````
